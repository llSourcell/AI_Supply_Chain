## 공급망이란 무엇입니까? 

![alt text](https://s9783.pcdn.co/wp-content/uploads/2017/06/Blog-CFO.png "Logo Title Text 1")

- 공급망이란 기업과 기업의 공급자 사이에서 특정 상품을 생산하고 분배하는 네트워크입니다. 공급망은 제품이나 서비스가 소비자에게 도달하는 일련의 과정을 상징합니다. 
- 최적화된 공급망은 비용을 낮춰주고 더 빠른 생산 사이클을 형성할 수 있도록 해줍니다. 따라서 공급망 관리는 아주 중요한 과정입니다. 
- 공급망은 물리적인 상품이 많은 세상에서 아주 중요한 요소입니다. 기업은 공급망을 최적화하고 개선하는 데 지속적으로 많은 비용과 노력을 투자하며, 이는 중단되지 않을 것입니다. 
- 5%의 수송 비용을 절약하면 큰 변화가 일어납니다. 
- 수송망 관리 시스템은 제조 과정, 특히 많은 부분으로 나뉘는 제조 과정의 복잡도와 비용을 줄일 수 있습니다. 
- 1 의류 제조업자는 원자재를 생산 라인으로 옮길 것입니다. 
- 2 그러면 제조업자는 기계를 가동하고 원자재를 사용하는 다른 일을 수행하는 데 노동 비용을 소모할 것입니다. 
- 3 제품이 생산되고 나면 소비자에게 팔릴 때까지 포장 및 저장될 것입니다.

## 공급망 문제

![alt text](https://image.slidesharecdn.com/scmseminar-120913101036-phpapp01/95/international-supply-chain-management-24-728.jpg?cb=1347531190 "Logo Title Text 1")

- 공급망을 계획하는 아주 비싼 인력을 많이 고용해야 합니다. 
- 공급망의 각 단계별로 
- 보통 다른 기능이나 파트너와 충돌을 일으킵니다. 
- 각각 국지적으로만 최적화되어 더 큰 기회를 놓치고 있습니다. 
- 상태 데이터로 작업하기 때문에 나쁜 결정으로 이어지고 있습니다. 
- 지나치게 단순화된 모델을 사용하여 실제 세상과 관련이 별로 없습니다.

## 스타트업은 AI를 사용하여 공급망을 어떻게 향상시킵니까? 

![alt text](https://vitalai.files.wordpress.com/2015/12/data-supply-chain-edv2015-hadfield-submitted-001.png?w=636 "Logo Title Text 1")

- 1 실시간 데이터에 접근합니다. 
- 오늘날의 많은 공급망은 오래 전 데이터를 토대로 계획을 세우지만, 이는 차선책에 불과하므로 의사 결정 과정이 다소 부실할 수 있습니다. 
- 2 커뮤니티 (다중 집단의) 데이터에 접근합니다 (기업 외부의 데이타는 매우 유용합니다) 
- 3 넓은 네트워크의 목적 함수를 지원합니다. 
- AI 솔루션은 공급망의 제한을 직면했을 때에도 소비자 주도의 목적 함수를 지원해야 합니다. 
- 4 의사 결정 과정은 이득을 증가시키는 방향으로 설정되어야 하며, 변경으로 인한 비용도 고려해야 합니다. 
- AI 도구는 의사를 결정할 때 이윤의 증가와 결정 변경으로 인한 비용간의 트레이드오프를 고려해야 합니다. 
- 5 의사 결정 과정은 지속적이고 자기 학습적이어야 하며 스스로를 모니터링 해야합니다. 

![alt text](http://www.datascienceassn.org/sites/default/files/users/user30/aa_0.jpg "Logo Title Text 1")

-  가변성과 지연 문제는 반복적인 문제이며, 실행 효율은 끊임없이 변화합니다. AI 시스템은 이 문제를 주기적으로가 보아서는 안되고, 끊임없이 주시해야 합니다. 
- 6 AI 엔진은 자동으로 의사를 결정해야 합니다. 
-  알고리즘이 똑똑한 결정만을 하는 게 아니라, 그것을 자동으로 수행할 수 있어야 유의미한 값을 얻을 수 있습니다. 
- 7 AI 엔진은 확장성이 있어야 합니다. 
- 큰 규모의 공급망은 수억개는 아니더라도 수백만개의 비축 장소가 있습니다. AI 솔루션은 반드시 큰 규모의 데이터에 대해서도 빠르고 현명한 판단을 내려야 합니다. 
- 8 사용자가 시스템과 맞물릴 방법이 있어야 합니다. 
- AI는 “어두운 상자” 속에서 작동하고 있어서는 안됩니다. 사용자 인터페이스는 사용자가 결정의 기준이나 파급 효과를 볼 수 있게 해야 하며, AI가 해결할 수 없는 이슈를 사용자가 이해할 수 있도록 해야 합니다.

### 물품 조달 가동을 위한 챗봇

![alt text](https://cdn-images-1.medium.com/max/1600/1*4L0robaTCDEuZnKtCpjioA.jpeg "Logo Title Text 1")

- 챗봇 역량의 자동화 및 증대를 통한 물품 조달 업무 간소화는 튼튼하고 똑똑한 데이터 집합을 요구합니다. 
- ‘procuebot’은  
- 1 간단한 대화중 공급자에게 말을 걸 수 있습니다.

- 3 구매 요청을 자동으로 배치합니다. 
- 4 물품 조달 기능 또는 공급자(들) 집합과 관련한 내부 질문을 조사하고 대답합니다. 
- 5 송장과 지불/구매 요청의 수신/filling?/문서화 (Smith 2016).

### 공급자 선택 및 공급자 관계 관리를 위한 머신 러닝과 예측 분석

![alt text](https://ars.els-cdn.com/content/image/1-s2.0-S1570868316300702-gr003.jpg "Logo Title Text 1")

- 공급자와 관련된 리스크는 국제적인 브랜드에게 족쇄가 되고 있습니다. 
- 공급자측의 운영이나 나쁜 홍보에서의 실수는 당신의 회사에 책임을 물을 수 있습니다. 
- 하지만 모든 공급자와의 상호 작용 도중에 공급자 선택 및 리스크 관리를 위한 최고의 가능한 시나리오를 가지고 있다면 어떨까요? 
- 공급자 평가, 회계 감사, 신용 평점과 같이 공급자 관계 관리에서 발생한 데이터 집합은 공급자와 관련된 이후의 결정에서 중요한 기반이 됩니다. 
- 머신 러닝과 이해하기 쉬운 알고리즘의 도움을 통해, 이러한 수동적인 데이터 수집은 능동적이 될 수 있습니다. 
- AI가 최선의 공급자를 말해줄 수 있습니다.
